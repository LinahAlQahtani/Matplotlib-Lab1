```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## Q1: Create two charts using list and dataframe.

```python
Clubs = ["Alhilal" , "Ittihad" , "Alshabab" , "Alnassr"]
Wins = [17 , 8 , 6 , 9]

plt.plot(Clubs , Wins )
plt.xlabel("Clubs")
plt.ylabel("Wins")
plt.title("Record of the League")

plt.show()

```

```python
x={ "Club": ["Alhilal" , "Ittihad" , "Alshabab" , "Alnassr"],
      "Wins": [17 , 8 , 6 , 9] }

y= pd.DataFrame(x)
plt.bar(y["Club"] , y["Wins"]);
plt.xlabel("Club")
plt.ylabel("Wins")
plt.title("Record of the League")

plt.show()
```

## Q2: Create pie chart with equal explode values.

```python
colors = ['#99ff99','#ff9999','#ffcc39' , '#66b3ff']
plt.pie(Wins, labels=Clubs, autopct='%1.1f%%' ,colors=colors, startangle=90 , shadow=True , radius=1 , explode=[0.2,0.2, 0.2 , 0.2])
plt.show()
```

## Q3: Create a scatter chart with circle marker, alpha = 0.4, and size = 50.

```python
plt.scatter(y["Club"] , y["Wins"] ,marker = 'o', s =50 , alpha = 0.4 );
plt.xlabel("Club")
plt.ylabel("Wins")
plt.title("Record of the League")
plt.show()
```

## Q4: Add the following to your chart:
- title placed in the left and has a red color
- x and y label with font size = 16

```python
plt.scatter(y["Club"] , y["Wins"]);
plt.xlabel("Club" , fontsize = 16)
plt.ylabel("Wins" , fontsize = 16)
plt.title("Record of the League" , loc='left' , color = 'red')
plt.show()
```

## Q5: Create three line charts with Legends at left and colors.

```python
plt.plot([10,9,8])
plt.plot([7,6,5])
plt.plot([4,3,2])


plt.legend(['line1' , 'line2', 'line3'] ,shadow = True ,loc= 2)
plt.show()
```

## Q6: Create subplots with three charts and figsize=20,10.

```python
plt.figure(figsize= [20,12])
plt.suptitle('Subplot' ) 

#first
plt.subplot(1, 3, 1)
plt.plot(Clubs , Wins )
plt.xlabel("Clubs")
plt.ylabel("Wins")
plt.title("Record of the League")
plt.show()  

#second 
plt.subplot(1, 3, 2)        
plt.scatter(y["Club"] , y["Wins"]);
plt.xlabel("Club")
plt.ylabel("Wins")
plt.title("Record of the League")
plt.show()

#third
plt.subplot(1, 3, 3)
plt.pie(Wins, labels=Clubs, autopct='%1.1f%%')
plt.show()
```
